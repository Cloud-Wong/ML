# 1. Logistic Regression

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#加载数据
data1 = pd.read_csv('ex2data1.txt',names = ['Exam1Score','Exam2Score','Admission'])
data1.head()
```

## 1.1 Visualizing the data

```python
pos = data1[data1['Admission']==1]
neg = data1[data1['Admission']==0]
fig,ax = plt.subplots(figsize=(8,6))
ax.set_title('Scatter plot of training data')
ax.scatter(pos['Exam1Score'],pos['Exam2Score'],color = 'black',marker = '+',label = 'Admitted')
ax.scatter(neg['Exam1Score'],neg['Exam2Score'],color = 'y',label = 'Not Admitted')
ax.set_xlabel('Exam 1 Score')
ax.set_ylabel('Exam 2 Score')
ax.legend()
```

## 1.2 Implementation
### 1.2.1 sigmoid function
logistic regression hypothesis is defined as:$$h_{\theta}(x) = g({\theta}^Tx)$$
where function g is the sigmoid function. The sigmoid function is defined as: $$g(z) = \frac{1}{1+e^{-z}}$$

```python
def sigmoid(z):
    z = np.array(z)
    return 1/(1+np.exp(-z))
```

### 1.2.2 Cost function and gradient
cost function in logistic regression is
$$J(\theta) = \frac{1}{m}\sum_{i=1}^{m}[-y^{(i)}log(h_{\theta}(x^{(i)}))-(1-y^{(i)})log(1-h_{\theta}(x^{(i)}))]$$
and the gradient of the cost is a vector of the same length as $\theta$ where the $j^{th}$
element (for j = 0,1,...,n) is defined as follows:
$$\frac{\partial{J(\theta)}}{\partial\theta_j} = \frac{1}{m}\sum_{i=1}^{m}(h_{\theta}(x^{(i)})-y^{(i)})x^{(i)}$$

```python
#return J and grad
def computer_cost(theta,*args):
    
    x,y = args
    m = x.shape[0]
    h = sigmoid(x@theta)
    first = np.log(h).T @ y 
    second = np.log(1-h).T @ (1-y)
    cost = -(first+second)/m
    return cost

def computer_grad(theta,*args):
    theta = theta.reshape(-1,1)
    x,y = args
    m = x.shape[0]
    hx = sigmoid(x@theta)
    grad = x.T @ (hx - y)/m
    return grad.ravel()
```

```python
#数据准备
x = data1.iloc[:,:2]
x.insert(0,'Ones',1)
x_test = x.values
y = data1.iloc[:,-1:]
y_test = y.values
theta = np.zeros((3,1))
J = computer_cost(theta,x_test,y_test)
J
```

### 1.2.3 Learning parameters using optimize

```python
import scipy.optimize as opt

#Newton-CG
res = opt.minimize(computer_cost,theta,args=(x_test, y_test),method='Newton-CG',jac = computer_grad)
#Nelder-Mead单纯形算法,由于它不使用任何梯度评估，因此可能需要较长时间才能找到最小值。
#res = opt.minimize(computer_cost,theta,args=(x_test, y_test),method='nelder-mead')
#res = opt.fmin_tnc(func=computer_cost, x0=theta, fprime=computer_grad, args=(x_test, y_test))
res.x
```

### 1.2.4 Evaluating logistic regression

```python
#预测函数
def pridect1(x,theta):
    if(sigmoid(x@theta)>=0.5):
        return 1
    return 0

final_theta = res.x.T
#For a student with an Exam 1 score of 45 and an Exam 2 score of 85
#you should expect to see an admission probability of 0.776.
sigmoid([1,45,85]@final_theta)
```

```python
#绘制决策边界
x = np.linspace(0,100)
parm = -final_theta/final_theta[2]
y = parm[0] + parm[1]*x
fig,ax = plt.subplots(figsize=(8,6))
ax.set_title('Training data with decision boundary')
ax.scatter(pos['Exam1Score'],pos['Exam2Score'],color = 'black',marker = '+',label = 'Admitted')
ax.scatter(neg['Exam1Score'],neg['Exam2Score'],color = 'y',label = 'Not Admitted')
ax.set_xlabel('Exam 1 Score')
ax.set_ylabel('Exam 2 Score')
ax.legend()
ax.plot(x,y)
plt.xlim(28,101)
plt.ylim(28,101)
```

# 2. Regularized logistic regression


```python
#载入数据
data2 = pd.read_csv('ex2data2.txt',names = ['Test1','Test2','Acception'])
data2.head()
```

## 2.1 Visualizing the data

```python
fig , ax = plt.subplots(figsize=(8,6))
pos = data2[data2['Acception'] == 1]
neg = data2[data2['Acception'] == 0]
ax.scatter(pos['Test1'],pos['Test2'],color = 'black',marker = '+')
ax.scatter(neg['Test1'],neg['Test2'],color = 'y')
ax.set_xlabel('Microchip Test 1')
ax.set_ylabel('Microchip Test 2')
ax.legend()
```

## 2.2 Feature mapping
$$mapFeature(x) = \begin{bmatrix}
1\\
{x_1}\\
{x_2}\\
{x_1^2}\\
{x_1x_2}\\
{x_2^2}\\
{x_1^3}\\
{\vdots}\\
{x_1x_2^5}\\
{x_2^6}\\
\end{bmatrix}$$


```python
#把特征映射成多项式
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(6)
x_test2 = (poly.fit_transform(data2.iloc[:,:2]))
pd.DataFrame(x_test2).head()
```

## 2.3 Cost function and gradient
cost:
$$J\left( \theta \right)=\frac{1}{m}\sum\limits_{i=1}^{m}{[-{{y}^{(i)}}\log \left( {{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)-\left( 1-{{y}^{(i)}} \right)\log \left( 1-{{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)]}+\frac{\lambda }{2m}\sum\limits_{j=1}^{n}{\theta _{j}^{2}}$$

gradient:
$$\frac{\partial J\left( \theta \right)}{\partial {{\theta }_{j}}}=\frac{1}{m}\sum\limits_{i=1}^{m}{\left( {{h}_{\theta }}\left( {{x}^{\left( i \right)}} \right)-{{y}^{\left( i \right)}} \right)} \text{ }\text{ for j}={0}$$


$$\frac{\partial J\left( \theta \right)}{\partial {{\theta }_{j}}}=\left( \frac{1}{m}\sum\limits_{i=1}^{m}{\left( {{h}_{\theta }}\left( {{x}^{\left( i \right)}} \right)-{{y}^{\left( i \right)}} \right)} \right)+\frac{\lambda }{m}{{\theta }_{j}}\text{ }\text{ for j}\ge \text{1}$$

```python
#数据准备
theta2 = np.zeros((x_test2.shape[1],1))
y_test2 = data2.iloc[:,-1:].values
```

```python
def cost_reg(theta,XX,yy,reg):
    m = XX.shape[0]
    J = 0
    grad = np.zeros((XX.shape[1],1))
    h = sigmoid(np.dot(XX,theta))
    theta_1 = theta[1:,:]
    J = -1*np.sum(yy*np.log(h) + (1-yy)*np.log((1-h))) / m + 0.5 * reg * theta_1.T.dot(theta_1) / m
    grad = XX.T.dot((h-yy)) / m 
    grad[1:,:] += reg*theta_1/m  #theta0 without reg
    return J,grad
```

```python
def bgd(X_train,y_train,theta,alpha = 0.1,iters = 5000,reg = 1):
    J_history = []
    for i in range(iters):
        cost,grad = cost_reg(theta,X_train,y_train,reg)
        theta = theta - alpha * grad
        J_history.append(float(cost))
    return theta,J_history
```

```python
#梯度下降
final_theta2,j_cost = bgd(x_test2,y_test2,theta2,reg=1)
plt.plot(j_cost)
plt.xlabel('iters')
plt.ylabel('j_cost')
plt.show()
```

## 2.4 Plotting the decision boundary

```python
#绘制散点图
fig , ax = plt.subplots(figsize=(8,6))
pos = data2[data2['Acception'] == 1]
neg = data2[data2['Acception'] == 0]
ax.scatter(pos['Test1'],pos['Test2'],color = 'black',marker = '+',label = 'y=1')
ax.scatter(neg['Test1'],neg['Test2'],color = 'y',label = 'y=0')
ax.set_xlabel('Microchip Test 1')
ax.set_ylabel('Microchip Test 2')
#绘制边界
theta_unfitting,_ = bgd(x_test2,y_test2,theta2,reg=100)
xx, yy = np.meshgrid(np.linspace(-0.75, 1),np.linspace(-0.75, 1))
z1 = sigmoid(poly.fit_transform(np.c_[xx.ravel(), yy.ravel()]).dot(final_theta2)) #boundary
z1 = z1.reshape(xx.shape)
z2 = sigmoid(poly.fit_transform(np.c_[xx.ravel(), yy.ravel()]).dot(theta_unfitting)) #boundary_unfitting
z2 = z2.reshape(xx.shape)
ax.contour(xx, yy, z1,[0.5],colors='b')
ax.contour(xx, yy, z2,[0.5],colors='r')
ax.plot([],[],color = 'b',label = r'$\lambda$ = 1')
ax.plot([],[],color = 'r',label = r'$\lambda$ = 100')
ax.legend()
```

```python
#预测函数
def pridect2(x):
    return (sigmoid(poly.fit_transform(x) @ final_theta2)>=0.5)
```

```python
#准确率
correctData = (data2.iloc[:,-1:]==pridect2(data2.iloc[:,:2]))
correct = len(correctData[correctData.Acception==True])/len(data2)
correct
```
